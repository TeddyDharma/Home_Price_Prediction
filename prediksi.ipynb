```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("harga_rumah.csv")
data
```

```python
jumlah_lantai = data['jml_lantai']
jumlah_ruangan = data['ruangan']
luas = data['luas (m2)']
harga = data['harga']
```

```python
all_data = [[lantai, ruangan, area] for lantai, ruangan, area in zip(jumlah_lantai, jumlah_ruangan, luas)]
for x in range(5): 
    print(all_data[x])
all_data = np.array(all_data)
harga = np.array(harga)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[3])
])
```

```python
model.compile( 
    loss='mae', 
    optimizer='sgd',
    metrics=['accuracy']
)
```

```python
model.fit(all_data, harga, epochs=500)
```
